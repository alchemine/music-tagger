```python
import os
from pathlib import Path
from tqdm import tqdm
import pandas as pd

from music_tagger.utils.config import STREAMING_DIR
from music_tagger.agents.tool import RecordSong
from music_tagger.music_tagger import MusicTagger


EXAMPLES = [
    # (
    #     "201129 네코제의 밤 윤하 - 우리, 다시 여기",
    #     RecordSong(title="우리, 다시 여기", artist="윤하"),
    # ),
    # (
    #     "【Official Music Video】アノマリー(Anomaly) feat.沖石",
    #     RecordSong(title="アノマリー(Anomaly)", artist="沖石"),
    # ),
]

tagger = MusicTagger(examples=EXAMPLES, example_type="system")
tagger.build()

# table = pd.DataFrame(columns=["name", "title", "artist"])
table = pd.DataFrame(columns=["name", *list(RecordSong.model_fields)])
for root, dirs, files in tqdm(list(reversed(list(os.walk(STREAMING_DIR))))):
    for file in files:
        name = Path(file).stem
        rst = tagger.run(file)
        rst["name"] = file
        rst["artist"] = rst["artist"] if rst.get("artist") else "#"
        table.loc[len(table)] = rst
        display(table.loc[[len(table) - 1]])
```
